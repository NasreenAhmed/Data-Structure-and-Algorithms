####Problem: 1
Find the factorial of given "n", using recursion.

```python
def factorial(n):

  #edge case when n = 0:
  if n == 0:
    return 1
  else:
    return factorial(n-1) * n
```

```python
if __name__ == '__main__':
  n = 7
  value = factorial(n)
  print(n,"!:",value)
```

####Problem: 2
Implement binary search, to find the data in a list, using recursion.

Notes: 
Binary search, is used to effectively locate a target value within a sorted sequence of n elements.This is among the most important computer algorithms, and it is  the reason that we so often store data in sorted order.

When the sequence is **unsorted**, the standard approach to search for a target  value is to loop to examine every element, until either finding the target or exhausting the data set.  This is known as **Sequential Search Algorithm**. This algorithm runs **$ ÎŸ(n)$**

```python
def binary_search(target,arr,low,high):
  """
  Return True if target is found in indicated portion of Python list. 
  The search only considers the portion from data[low] to data[high]
  """

  if low > high:
    return False   # Interval is empty no match
  else:
    mid = (low + high)// 2
    if target == arr[mid]:
      return True
    elif target > arr[mid] :
      # recur on the portion right of the middle
      return binary_search(target,arr,mid+1,high)
    else:
      # recur on the portion left of the middle
      return binary_search(target,arr,low,mid-1)
```

```python
arr = [2,4,5,7,8,12,14,17,19,22,25,27,28,33,37]

target= 22
low = 0
high = len(arr)-1
result = binary_search(target,arr,low,high)

if result is True:
  print("The given target data is present in the given list" )
else:
  print("The given number target data is not present in the list")
```

A Recursive approach to Ruler drawing 

The english ruler pattern is a simple example  of a fractal, that is, a shape that has a self-recursive at various stages of maginification.
Ignoring the lines containing 0,1,2. let us consider how to draw  the sequence of ticks lying between the lines. The central tick between 0 and 1, had length 3. Observe that the two patterns of ticks above and below this tick are identical and has a central tick of length 2.

In general, an interval with a central tick of length $L \geq 1$ is composed of :      
- An interval with central tick length $L-1$
- A single tick of length $L$.
- An interval with a central tick of length $L-1$

Our implementation consists of three functions:     

1. Main Function :  `draw_ruler` : Manages the construction of the entire ruler. Its arguments specify the the total number of inches/cms in the ruler and the major tick.

2. Utility Function : `draw_line`: Draws a single tick with a specified number of dahses.

3. Recursive function:`draw_interval` : This function draws the sequence of minor ticks within some interval,based upon the length of the interval's central tick.

```python
def draw_line(tick_length,tick_label=''):
  """
  Draw one line with given tick length( followed by optional label)
  """
  line = '-' * tick_length
  if tick_label:
    line+=' ' + tick_label
  print(line)

def draw_interval(center_length):
  """
  Draw tick interval based upon a central tick length
  """
  if center_length >0:                       #stops when length drops to zero
    draw_interval(center_length - 1)          #recursively draw top ticks
    draw_line(center_length)                 #Draw central tick
    draw_interval(center_length - 1)         # recursively draws bottom ticks.

def draw_ruler(num_inches,major_length):
  """
  Draw english ruler with given number of inches, major tick length
  """
  draw_line(major_length,'0')               #draw inch 0 line
  for j in range(1,num_inches+1):
    draw_interval(major_length-1)           #draw interior ticks for inch
    draw_line(major_length, str(j))         #draw inch j line and label
```

```python
draw_ruler(2,4)
```

### Problem : 4

Reversing a list using recursion

```python
def reverse(arr,start,stop):
  if start < stop:
    arr[start],arr[stop] = arr[stop], arr[start]
    reverse(arr,start+1,stop-1)
```

```python
arr = [4,3,6,2,4,8,9,5]
start = 0 
stop = len(arr)-1
reverse(arr,start,stop)
print("Reversed arr",arr)
```

### Problem : 5

Summing the elements of a sequence recursively

```python
def linear_sum(arr):
  n = len(arr)
  if n == 0:
    return 0
  else:
   return linear_sum(arr[0:n-2]) + arr[n-1]
```

```python
arr = [4,3,6,2,8,9,3,2,8,5,1,7,2,8,3,7]
sum = linear_sum(arr)
print("The sum of the elements of the sequence", sum)
```

### Problem : 6
 Recursive Algorithms for computing power

We consider the problem of raising a a number $x$ to an arbitrary integer $n$. That is we wish to compute the **power function**, *defined* as **$pow(x,n)=x^n$**. We will use two different recursive formulas for the problem.

1. Naive implementaton of power function
2. Advanced implementation of the power function.

```python
#Naive Implementation, similar to Factorial : Our first recursive algorithm.
def pow_naive(x,n):

  if n == 0:
    return 1
  else:
    return pow(x,n-1) * x
```

```python
def pow(x,n):

  if n == 0:
    return 1
  else:
    partial = pow(x,n//2)
    print("partial",partial)
    result = partial * partial 
    if n%2 == 1:
      result *= x
    return result
```

```python
x = 2
n = 13
result = pow(x,n)
print(result)
```

###Problem : 7 

**Binary Recursion** :  When a function makes two recursive calls, we say that it uses binary recursion. Computing the sum of zero or one elements is trivial. With two or more elements, we can recursively compute  the sum of the first half, and the sum of the second half  and add the two sums together. 

```python
def binary_summation(arr,low,high):

  if low > high:       # 0 elements in the list
    return 0
  elif low == high:
    return arr[low]
  else:
    mid = (low+high)//2
    bottom = binary_summation(arr,low,mid)
    top = binary_summation(arr,mid+1,high)
    return bottom + top 
```

```python
arr = [6,44,33,22,11,66,43,21]
low = 0
high = len(arr) - 1
binary_summation(arr,low,high)
```

### Problem : 8

Describe a recursive algorithm for finding the maximum element in a sequence, $S$, of $n$ elements. What is your running time and space usage? 

```python
def bin_max(arr,start,stop):

  if start > stop:        # the array/list is empty
    return 0
  elif start == stop:
    return arr[start]
  else:
    mid = (start + stop) //2
    first_half = bin_max(arr,start,mid)
    second_half = bin_max(arr,mid+1,stop)
    if first_half >= second_half:
      return first_half
    else:
      return second_half
```

```python
arr = [1,2,1,0,50,9,8,30,40,0,3,2,4,1,6,80]
start = 0
stop = len(arr) - 1
result = bin_max(arr,start,stop)
print("The maximum element in the sequence S of n elements :",result)
```

### Problem : 9

Draw the recursion trace for the computation of **$pow(2,18)$**, using the repeated squaring algorithm

```python
def pow(x,n):

  if n == 0:
    return 1
  else:
    partial = pow(x,n//2)
    result = partial * partial 
    if n % 2 == 1:
      result = partial * partial * x
    return result 
```

### Problem : 10

Draw the recursion trace for the execution of the function reverse $(S,0,5)$ on $S=[4,3,6,2,6]$

```python
def reverse(S,start,stop):

  if start > stop:      #the list is empty 
    return False 
  else:
    S[start],S[stop] = S[stop], S[start]
    return reverse(S,start+1,stop-1)
```

### Problem : 11
Write a short recursive Python function  that finds the minimum and maximum values  in a sequence without using any loops.

```python
def min_max(S,start,stop):

  if start > stop:                # the given list is empty 
    return 0
  elif start == stop-1:           # When we have only two elements left in the sub-array
    if S[start] > S[stop]:        # Make a comparison among the two, and designate min and max among them 
      min = S[stop]
      max = S[start]
    else:
      min = S[start]
      max = S[stop]
    return min,max
  else:
    mid = (start + stop)//2        # calculat the mid 
    min1,max1 = min_max(S,start,mid)
    min2,max2 = min_max(S,mid,stop)

    if min1 < min2:               # when both the recursion calls are done, compare the min against min, and max against max
      min = min1                  # and return the min and max as per the conditions satisfied.
    else:
      min = min2
    
    if max1 > max2:
      max = max1
    else:
      max = max2
    return min,max
```

### Problem : 12

Give a recursive algorithm to compute the product of two positive integers ,$m$ and $n$, using only adddition and subtraction 

```python
def prod(m,n):

  if n == 0:
    return 0
  else:
    if m > n:                 # if m > n: means logically, we repeat (n times): lesser number of recursion call
      return m + prod(m,n-1)
    else:                      #else: interchange the values of m and n, to facilitate, lower number of recursion calls.
      return n + prod(n,m-1)
```

```python
m = 100
n = 10
multi = prod(m,n)
print("The value of m*n :",multi)
```
