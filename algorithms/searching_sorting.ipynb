# <font color = 'darkblue'>Searching </font>

<font color = 'blue'>Searching</font> is the process of selecting particular information from a collection of data based on specific criteria.

- The search operation can be performed on many different data structures. the **Sequence Search**, involves finding an item within a sequence  using s **search key** to identify the specific item.
- A **key**  is a unique value used  to identify data elements of a collection.
- In collection containing simple types such as integers or real, the value themselves are keys.
- For collections of complex type, a specific data component has to be identified as the key.In some instances, a key may consist of multiple components, which is also known as **compound key**.

## <font color ='blue'>1. Linear Search</font>

The simplest solution to the sequence search problem is the sequential or **linear search** algorithm. This technique iterates over the sequence, one item at a time, until the specific item is found or all items have been examined.


```python
#implementation of Linear search : Worst case complexity : O(n)
def linear_search(arr,target):
  n = len(arr)
   for i in range(n):
     # if the target is in the ith item return true.
     if arr[i] == target:
       return True 
   return False   #if not found return false.
```

### <font color='blue'>Searching a Sorted Sequence</font>

A linear search can also be performed on a sorted sequence, which is a sequence containing values in a specific order. 

A linear search on a sorted sequence works in the same fashion as that for the unsorted sequence, with one exception. It’s possible to terminate the search early when the value is not in the sequence instead of always having to perform a complete traversal.

The only modification to the earlier version is the inclusion of a test to deter- mine if the current item within the sequence is larger than the target value. If a larger value is encountered, the loop terminates and False is returned.

```python
#Implementation of Linear Search on a Sorted sequence. Still Worst case complexity is O(n)
def sorted_linear_search(arr,target):
  n = len(arr)
   for i in range(n):
     # if the target is in the ith item return true.
     if arr[i] == target:
       return True 
    elif arr[i] > target:   # if the ith element is greater than the target, then it is not in the sequence. 
      return False
   return False   #if not found return false.
```

## <font color='blue'>2. Binary Search</font>

The linear search algorithm for a sorted sequence produced a slight improvement over the linear search with an unsorted sequence, but both have a linear time-complexity in the worst case. To improve the search time for a sorted sequence, we can modify the search technique itself.

**Algorithm Description**

The algorithm starts by examining the middle of the sorted sequence, resulting  in one of the three possible conditions; 
- the middle item is the target value
- the target value is less than the middle item
- the target value is more than the middle item 

Since the sequence is ordered, we can eliminate half the values in the list when the target value is not found at the middle position.

**Implementation**

The variables **low and high** are used to mark the range of elements in the sequence current under consideration.  When the search begins, this range is the entire sequence since the target  item can be anywhere within the sequence. The first step in each iteration is to determine the mid point  of the sequence. 

```python
#Implementation of the Binary Search on a Sorted Sequence . Worst Case Complexity : O(log n)

def sorted_binary_search(arr,target):

  low = 0
  high = len(arr) - 1 
  while low <= high:
      # We get the loop running until the condition remains stisfied. 
      # Find the midpoint of the sequence
      mid = (low + high)//2 
      # Does the midpoint contain the target ? 
      if target == arr[mid]:
        print(" The target is found at position :",mid, "in the given array")
        return True
       # Does the target precede the midpoint ?
      elif target < arr[mid]:
        high = mid - 1
        # or does it follow the midpoint ?
      else:
        low = mid + 1
  
  # If the sequence cannot be subdivided further we're done.
  return False
```

### Run Time Analysis 

To evaluate the efficiency of the the binary search algorithm, assume the sorted sequence contains n items. <font color='blue'>We need to determine the maximum number of times the while loop is executed.</font> 
The worst case occurs when the target value is not in the sequence, the same as for the linear search. The difference with the binary search is that not every item in the sequence has to be examined before determining the target is not in the sequence, even in the worst case. Since the sequence is sorted, each iteration of the loop can eliminate from consideration half of the remaining values. when the input size is repeatedly re- duced by half during each iteration of a loop, there will be log n iterations in the worst case. Thus, the binary search algorithm has a worst case time-complexity of O(log n), which is more efficient than the linear search.

# <font color = 'darkblue'>Sorting </font>

Sorting is the process of arranging or ordering a collection of items  such that each item and its successor satisfy a prescribed relationship. The items can be simple  values,  such as intergers or reals. or more complex types, such as student records or dictionary entries.  In either case, the ordering, of the items is based  on the value of a **sort key**. In this section, we present three basic sorting algorithms, all of which  can be applied to data stored in a mutable sequence such as an array or list. 

## <font color='blue'>1. Bubble Sort </font>

A simple solution to the sorting problem is the **Bubble Sort Algorithm** , which  rearranges the values by iterating over the list multiple times, causing larger values to buddle  to the top of or end of the list.
To illustrate how the bubble sort algorithm works, suppose we have four playing cards (all of the same suit) that we want to order from smallest to largest face value. We start by laying the cards out face up on a table as shown here:

def bubble_sort(arr)⁉

```python
# Sorts a Sequence in ascending order using the bubble sort algorithm
def bubble_sort(arr):

  i = 0
  n = len(arr)
  # perform n-1 bubble operations on the sequence.
  while i < n:

   # Bubble the larget item to the end.
    for j in range(0,n-i-1):

      if arr[j] > arr[j+1]:    #swap j and j+1 terms
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
      print("Outer loop num:",i,"Inner loop num :",j,"Partial sorted Array :",arr)

    i +=1

  return arr  
```

```python
#arr = [8,5,9,3]
arr = [10,2,18,31,13,5,64,29]
result = bubble_sort(arr)
```

- The efficiency of the bubble sort algorithm only depends on the number of keys in the array and is independent of the specific values and the initial arrangement of those values.

- To determine the efficiency, we must determine the total number of iterations performed by the inner loop for a sequence containing n values. The outer loop is executed $n − 1$ times since the algorithm makes $n − 1$ passes over the sequence. The number of iterations for the inner loop is not fixed, but depends on the current iteration of the outer loop. On the first pass over the sequence, the inner loop executes $n − 1$ times; on the second pass, $n − 2$ times; on the third, $n − 3$ times, and so on until it executes once on the last pass. 

- The total number of iterations for the inner loop will be the sum of the first $n − 1$ integers, which equals:      

\begin{align}
\frac{(n-1)(n-1+1)}{2}=\frac{n(n-1)}{2} - n = \frac{1}{2} n^2 +\frac{1}{2} n
\end{align}

resulting in a run time of $O(n^2)$.

## <font color='blue'>2. Improved Bubble Sort </font>

- The bubble sort algorithm can be improved by having it terminate early and not require it to perform all $n^2$ iterations when the sequence is in sorted order. 

- We can determine the sequence  is in sorted order when no swaps are performed by the `if` statement within the inner loop. At that point, the function can return immediately without completing the remanining iterations.

```python
# Sorts a Sequence in ascending order using the Improved bubble sort algorithm
def improved_bubble_sort(arr):

  i = 0
  n = len(arr)
  count = 0
  iter = 0
  # perform n-1 bubble operations on the sequence.
  while i < n:
    count = 0
   # Bubble the larget item to the end.
    for j in range(0,n-i-1):
      iter +=1
      print("i",i,"Iteration : ",iter,"|","Array :",arr)
      if arr[j] > arr[j+1]:    #swap j and j+1 terms
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
        count +=1


    print("count",count)
    if count == 0:
      print("The arr is already sorted, hence further iteration of outer while loop aren't needed")
      return arr     

    i +=1

  return arr  
```

## <font color='blue'>3. Selection Sort </font>

A second sorting algorithm, which improves on the bubble sort and works in a fashion similar to what a human may use a sort of  list of values, is known as **selection sort**.

We can again use the set of playing cards to illustrate the algorithm and start by placing five cards face up on a table that are to be sorted in ascending order.

The selection sort, which improves on the bubble sort, makes multiple passes over the sequence, but unlike the bubble sort, it only makes a single swap after each pass.

```python
# Sorts a sequence in ascending order using selection sort algorithm.
def selection_sort(arr):
  
  n = len(arr)

  for i in range(0,n):
    # Assume the ith element is the smallest.
    small = i
    #print("Before the traversal small : i",small)
    # determine if any other value contain the smaller value
    for j in range(i+1,n):
      if arr[j] < arr[small]:
        small = j
    #print("After the traversal small : j",small)
    #Swap the ith value  and small index value if the smallest value is not already in 
    # its proper position. Some impementations omit testing the condition and always
    # swaps the values.
    if small != i : 
      temp = arr[i]
      arr[i] = arr[small]
      arr[small] = temp
    print("i",i,"Arr:",arr)
```

```python
arr = [10,51,2,18,4,31,13,5,23,64,29]
result = selection_sort(arr)
```

- The process starts by finding the smallest value in the sequence and swaps it with the value in the first position of the sequence. 
- The second smallest value is then found and swapped with the value in the second position. 
- This process continues positioning each successive value by selecting them from those not yet sorted and swapping with the values in the respective positions.

**The selection sort, which makes $n − 1$ passes over the array to reposition $n − 1$ values, is also $O(n^2)$. The difference between the selection and bubble sorts is that the selection sort reduces the number of swaps required to sort the list to $O(n)$**.

## <font color='blue'>4. Insertion Sort </font>


```python
#Naive Implementation of Insertion Sort.
def insertion_sort(arr):

  n = len(arr)
  for i in range(0,n):

    ele = arr[i]

    for j in range(0,i):

      if arr[j] > ele:
        temp = arr[i]
        arr[i] = arr[j]
        arr[j] = temp
    print("Arr",arr)
```

```python
arr = [3,8,5,9,6]
result = insertion_sort(arr)
```

<font color='blue'>The insertion sort maintains a collection of sorted items and a collection of items to be sorted.</font>>
- In the playing card analogy, the deck represents the collection to be sorted and the cards in our hand represent those already sorted.

- When implementing insertion sort in a program, the algorithm maintains,both the sorted and unsorted collection within the same sequence structure. 
- The algorithm keeps the list of sorted values at the front of the sequence and picks the next unsorted value from the first of those yet to be positioned.
- **To position the next item, the current spot within the sequence of sorted values is found by performing a search.**.

```python
# Book Implementation : Implementing Insertion Sort Algorithm
# Sorts a sequence in ascending order using the insertion sort.
def insertionSort(theSeq):
  n = len(theSeq)
  # Starts with the first item as the only entry. 
  for i in range(1,n):
    # Save the value to be positioned.
    value = theSeq[i]
    print("i",i,"Value",value)
    # Find the position where the value  fits in the  ordered part of the list. 
    pos = i
    while pos > 0 and value < theSeq[pos-1]:
      # Shift the items to the right during the search.
      theSeq[pos] = theSeq[pos-1]
      print("Inside the while :", "Arr:", theSeq)
      pos-=1
    
    # Put the saved value in the open slot.
    theSeq[pos] = value
    print("i",i,"Arr : ",theSeq,"\n")
```

```python
arr = [10,51,2,18,4,31,13,5,23,64,29]
result = insertionSort(arr)
```

- The insertionSort() function starts by assuming the first item is in its proper position. Next, an iteration is performed over the remaining items so each value can be inserted into its proper position within the sorted portion of the sequence. 
- The ordered portion of the sequence is at the front while those yet to be inserted are at the end. 
- The **i loop index variable marks the separation point** between the two parts.
- **The inner loop is used to find the insertion point within the sorted sequence and at the same time, shifts the items down to make room for the next item.**
- Thus, the inner loop starts from the end of the sorted subsequence and works its way to the front. After finding the proper position, the item is inserted.

## <font color='blue'>1. Maintaining a Sorted list</font>

- To maintain a sorted list in real time, new items must be inserted into their proper position. The new items cannot simply be appended at the end of the list as they may be out of order. 

- Instead, we must locate the proper position within the list and use the `insert()` method to insert it into the dedicated position.

- To find the position of a new item within a sorted list,a modified version of the **binary search algorithm** can be used. The binary search uses a divide and conquer strategy to reduce the number of items that must be examined to find the target item or to determine the target is not in the list. 

- Instead of returning `True` or `False` indicating the existence of a  value, we can modify the algorithm to return the index position of the target if it's  actually in the list  or where the value should  be placed if it were inserted into the list. 





```python
# Modified version of the binary search that returns the index within
# a sorted sequence indicating where the target should be located.
def findSortedPosition( theList, target ): 
  low = 0
  high = len(theList) - 1
  while low <= high :
    mid = (high + low) // 2
    if theList[mid] == target :
      return mid
    elif target < theList[mid] :
      high = mid - 1 
    else :
  # Index of the target.
      low = mid + 1

  return low # Index where the target value should be.
```

## <font color='blue'>2. Merging Sorted Lists</font>

```python
# Merges two sorted  lists to create and return a new sorted list.

def mergeSortedLists(listA,listB):

  #create the new list and  initialize the list markers.
  newList = []
  a = 0 
  b = 0

  # Merge the two lists together until one is empty.
  while a < len(listA) and b < len(listB):
    if listA[a] < listB[b]:
      newList.append(listA[a])
      a +=1
    else:
      newList.append(listB[b])
      b+=1
  
  #If listA contains more elements than listB.
  while a < len(listA):
    newList.append(listA[a])
    a+=1

    #If listB contains more elements than listB.
  while b < len(listB):
    newList.append(listB[b])
    b+=1    

  return newList
```
